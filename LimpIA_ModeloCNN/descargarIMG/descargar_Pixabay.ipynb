```python
import requests
import os

# ⚙️ CONFIGURACIÓN
API_KEY = '49546674-'
QUERY = 'people at the beach'
NUM_IMAGES = 20
OUTPUT_DIR = './data/neutral'
PER_PAGE = 10  # máximo permitido por Pixabay

# Crear carpeta si no existe
os.makedirs(OUTPUT_DIR, exist_ok=True)

def download_images():
    downloaded = 0
    page = 1

    while downloaded < NUM_IMAGES:
        url = (
            f'https://pixabay.com/api/?key={API_KEY}&q={QUERY.replace(" ", "+")}'
            f'&image_type=photo&per_page={PER_PAGE}&page={page}&safesearch=true'
        )

        response = requests.get(url)
        data = response.json()

        if 'hits' not in data or not data['hits']:
            print("⚠️ No hay más imágenes disponibles. Fin de resultados.")
            break

        for hit in data['hits']:
            img_url = hit['largeImageURL']
            img_id = hit['id']
            img_path = f'{OUTPUT_DIR}/{img_id}.jpg'
            downloaded += 1

            # Evitar duplicados
            if os.path.exists(img_path):
                print(f'❌{downloaded}.-Saltada (ya existe): {img_id}')
                continue

            img_data = requests.get(img_url).content
            with open(img_path, 'wb') as f:
                f.write(img_data)

            print(f'{downloaded}.- Descargada {img_id}')

            if downloaded >= NUM_IMAGES:
                break

        page += 1

    print(f'\n✅ {downloaded} imágenes únicas descargadas en: {OUTPUT_DIR}')

download_images()
```
