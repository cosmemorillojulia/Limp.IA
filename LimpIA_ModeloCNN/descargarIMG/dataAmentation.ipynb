```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img, array_to_img

# ⚙️ CONFIGURACIÓN
INPUT_DIR = './data/neutral'           # Carpeta con tus imágenes originales
OUTPUT_DIR = './data/aumented'  # Carpeta donde se guardarán las aumentadas
AUGMENTATIONS_PER_IMAGE = 3               # Cuántas versiones nuevas por imagen

# Crear carpeta de salida si no existe
os.makedirs(OUTPUT_DIR, exist_ok=True)

# 💡 Generador de aumentos
datagen = ImageDataGenerator(
    horizontal_flip=True,
    rotation_range=25,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=10,
    zoom_range=0.1,
    brightness_range=(0.8, 1.2),
    fill_mode='nearest',
    rescale=1./255
)

# 📸 Aplicar aumentos a cada imagen del directorio
for filename in os.listdir(INPUT_DIR):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(INPUT_DIR, filename)
        img = load_img(img_path)
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)

        # Generar imágenes aumentadas
        i = 0
        for batch in datagen.flow(x, batch_size=1):
            new_filename = f'{os.path.splitext(filename)[0]}_aug_{i}.jpg'
            if i == 0:
                out_path = os.path.join(OUTPUT_DIR, filename)
            else:
                out_path = os.path.join(OUTPUT_DIR, new_filename)
                
            array_to_img(batch[0] * 255).save(out_path)
            i += 1
            if i >= AUGMENTATIONS_PER_IMAGE:
                break

print(f'\n✅ Data augmentation completado. Imágenes guardadas en: {OUTPUT_DIR}')
```
