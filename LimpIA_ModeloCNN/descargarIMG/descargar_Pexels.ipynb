```python
import requests
import os

# Configuración
API_KEY = ''
QUERY = 'personas en la playa'
NUM_IMAGES = 20
OUTPUT_DIR = './data/neutral'

# Crear carpeta si no existe
os.makedirs(OUTPUT_DIR, exist_ok=True)

headers = {
    'Authorization': API_KEY
}

def download_images():
    per_page = 15
    downloaded = 0
    page = 1

    while downloaded < NUM_IMAGES:
        url = f'https://api.pexels.com/v1/search?query={QUERY}&per_page={per_page}&page={page}'
        response = requests.get(url, headers=headers)
        data = response.json()

        if 'photos' not in data or not data['photos']:
            print("⚠️ No hay más imágenes disponibles. Fin de resultados.")
            break

        for photo in data['photos']:
            img_url = photo['src']['large']
            img_id = photo['id']
            img_path = f'{OUTPUT_DIR}/{img_id}.jpg'
            downloaded += 1

            # Si la imagen ya existe, la saltamos
            if os.path.exists(img_path):
                print(f'❌{downloaded}.- Saltada (ya existe): {img_id}')
                continue

            img_data = requests.get(img_url).content
            with open(img_path, 'wb') as f:
                f.write(img_data)
            

            print(f'{downloaded}.- Descargada {img_id}')

            if downloaded >= NUM_IMAGES:
                break

        page += 1

    print(f'\n✅ {downloaded} imágenes únicas descargadas en: {OUTPUT_DIR}')

download_images()
```
